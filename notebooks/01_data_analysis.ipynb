```python
import pandas as pd
```

```python
#Load datasets

titles = pd.read_csv("../data/titles.csv")
profiles = pd.read_csv("../data/profiles.csv")
results = pd.read_csv("../data/results.csv")
```

```python
#Display basic info

print("=== Titles Dataset ===")
print(titles.info())
print(titles.head())

print("\n=== Profiles Dataset ===")
print(profiles.info())
print(profiles.head())

print("\n=== Results Dataset ===")
print(results.info())
print(results.head())
```

```python
#Quick statistics

print("\n=== Titles Stats ===")
print(titles.describe(include='all'))

print("\n=== Profiles Stats ===")
print(profiles.describe(include='all'))

print("\n=== Results Stats ===")
print(results.describe(include='all'))
```

```python
#Checking for missing values

print("\n=== Missing Values ===")
print("Titles missing:\n", titles.isnull().sum())
print("Profiles missing:\n", profiles.isnull().sum())
print("Results missing:\n", results.isnull().sum())
```
